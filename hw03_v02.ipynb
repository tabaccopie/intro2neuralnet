```python
import pandas as pd
import numpy as np

from sklearn.datasets import fetch_california_housing # данный датасет взял вместо устаревшего boston_house
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, r2_score as r2

import tensorflow as tf
from tensorflow import keras

import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = fetch_california_housing(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=24)
```

```python
X_train.shape
```

```python
model = keras.Sequential([
  keras.layers.Dense(64, activation='relu',
                      input_shape=(X_train.shape[1],)),
  keras.layers.Dense(32, activation='relu'),
  keras.layers.Dense(1)
])
```

```python
model.compile(loss='mse',
              optimizer=keras.optimizers.Nadam(learning_rate=0.001),
              metrics=['mae'])
```

```python
model.fit(
    X_train,
    y_train,
    epochs=5,
    verbose = 0
)
```

```python
model.evaluate(X_train, y_train)
```

```python
results = []

for n in range(10):
  for metrics in ['mean_absolute_error', 'mean_squared_error']:
      for loss in ['mse', 'mae']:
        model = keras.Sequential([
        keras.layers.Dense(64, activation='relu',input_shape=(X_train.shape[1],)),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dense(1)
        ])

        result = {}

        model.compile(optimizer=keras.optimizers.Nadam(learning_rate=0.001),
                      loss=loss,
                      metrics=metrics)
            
        model.fit(X_train, y_train,
                  epochs=50, verbose = 0)
        
        predictions = model.predict(X_test)

        result['n'] = n
        result['metrics'] = metrics
        result['loss'] = loss
        result['r2'] = round(r2(y_test, predictions.copy().flatten()),5)
        result['rmse'] = round(np.sqrt(mse(y_test, predictions.copy().flatten())),5)
        result['mse'] = round(mse(y_test, predictions.copy().flatten()),5)
        result['mae'] = round(mae(y_test, predictions.copy().flatten()),5)


        results.append(result)


result = pd.DataFrame(results)
result
```
